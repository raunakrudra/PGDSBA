```python
Predicted_data['loan_status_pred']=Predicted_data['loan_status_pred'].astype(str)
Predicted_data['loan_status_pred'] = Predicted_data['loan_status_pred'].str.replace('0','Fully Paid')
Predicted_data['loan_status_pred'] = Predicted_data['loan_status_pred'].str.replace('1','Charged Off')
```

```python
Predicted_data.head()
```

```python
Predicted_data['Score']=0
```

```python
prob_min=0
prob_max=1
score_min=300
score_max=800
int_rate_min=2
int_rate_max=22
```

```python
Predicted_data['Score']=round(((1-Predicted_data['prob'])-prob_min)/(prob_max-prob_min)*(score_max-score_min)+score_min)
```

```python
#300-500 high_risk
#500-650 Medium risk
#650-800 low_risk
# create a list of our conditions
conditions = [
    (Predicted_data['Score'] <= 500),
    (Predicted_data['Score'] > 500) & (Predicted_data['Score'] <= 650),
    (Predicted_data['Score'] > 650)
    ]

# create a list of the values we want to assign for each condition
values = ['high_risk', 'Medium_risk', 'low_risk']

# create a new column and use np.select to assign values to it using our lists as arguments
Predicted_data['Risk_category'] = np.select(conditions, values)
```

```python
Predicted_data['int_rate_pred']=0
```

```python
Predicted_data['int_rate_pred']=round((Predicted_data['prob']-prob_min)/(prob_max-prob_min)*(int_rate_max-int_rate_min)+int_rate_min)
```
